```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
from matplotlib.pyplot import MultipleLocator
```

```python
data = pd.read_csv('chinaco2.csv')
xdata = data.loc[:,'Date']
ydata1 = data.loc[:,'ch20']
ydata2 = data.loc[:,'ch19']
```

Copy the picture

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(18,9))
plt.plot(xdata,ydata1)
plt.plot(xdata,ydata2)
line2, = plt.plot(xdata,ydata2,'#000093',label="2019",linewidth=8)
line1, = plt.plot(xdata,ydata1,'#F75000',label="2020",linewidth=8)
plt.text(45, 32, '2019',c='#000079',fontsize=25,fontweight='heavy')
plt.text(270, 32, '2020',c='#F75000',fontsize=25,fontweight='heavy')
plt.legend(bbox_to_anchor=(-0.44, -0.18, 0.99, 0.1),ncol=2)
plt.xticks(range(0,280,30),fontsize = 20,c='gray')
plt.ylabel('MtCO2 per day',c='gray',fontsize=25,fontweight='heavy')
plt.title("\nChina",fontsize=30,fontweight='heavy')
ax = plt.gca()
y_major_locator=MultipleLocator(2)
ax.yaxis.set_major_locator(y_major_locator)
plt.ylim(16,36)
plt.yticks(fontsize = 20,c='gray')
plt.axhline(y=36,c='gray',lw=0.5)
plt.axhline(y=34,c='gray',lw=0.5)
plt.axhline(y=32,c='gray',lw=0.5)
plt.axhline(y=30,c='gray',lw=0.5)
plt.axhline(y=28,c='gray',lw=0.5)
plt.axhline(y=26,c='gray',lw=0.5)
plt.axhline(y=24,c='gray',lw=0.5)
plt.axhline(y=22,c='gray',lw=0.5)
plt.axhline(y=20,c='gray',lw=0.5)
plt.axhline(y=18,c='gray',lw=0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('gray')
plt.savefig('copypic.png')
plt.show()
```
